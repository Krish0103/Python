```python
import yfinance as yf
```

```python
ticker = input("Enter the stock symbol: ")

df = yf.download(ticker, start = "2023-01-01", end = "2024-01-01", interval="1d")
df.head()
```

```python
daily_close = df["Close"]
daily_close.head(20)
```

```python
plt.figure(figsize=(10,6))
plt.plot(df["Close"])
plt.title("analysis")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.show()
```

```python
df["7_day_MA"] = df["Close"].rolling(window=7).mean()

df.head(20)
```

```python
df["30_day_MA"] = df["Close"].rolling(window=30).mean()

print(df)
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Close"], label = "Daily clase price")
plt.plot(df["7_day_MA"], label = "7_day_MA")
plt.plot(df["30_day_MA"], label = "30_day_MA")
plt.title(f" {ticker} Stock Price Time_Series")
plt.legend()
plt.tight_layout()
plt.show()
```
