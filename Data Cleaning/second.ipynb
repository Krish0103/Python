```python
import numpy as py
import pandas as pd
ds = pd.read_csv("bank_transactions.csv")
```

```python
ds.info()
```

```python
#Convert date column into datetime format
ds["date"] = pd.to_datetime(ds["date"])
```

```python
# before analysis we must ensure the data is logically correct

errors = []
if ds["amount"].min() <= 0:
  errors.append("Invalid transaction ant found")

if ds.duplicated("transaction_id").any():
  errors.append("Duplicate transaction id found")

if ds["balance"].min() <= 0:
  errors.append("Negative Balance")
```

```python
if errors:
  print("Data Validation failed")
  
  for i in errors:
    print("-", i)
else:
  print("Data Validation Passed")
```

```python
# Calculate total transaction amount per day
daily_summary = ds.set_index("date").resample("D")["amount"].sum()
print(daily_summary)
```

```python
monthly_summary = ds.set_index("date").resample("ME")["amount"].sum()
print(monthly_summary)
```

```python
# calculate the monthly summary
monthly_summary = ds.set_index("date").resample("ME")["amount"].mean()
print("\n Monthly average transaction amount: ")
print(monthly_summary)
```

```python
# Detect suspicious Transactions
suspicious_transactions = ds[ds["amount"]>8000]

suspicious_transactions_count = len(suspicious_transactions)

print("Total Suspicious transcations found:",suspicious_transactions_count)
```

```python
suspicious_transactions.to_excel("suspicious Transactions.csv", index = False)
```
